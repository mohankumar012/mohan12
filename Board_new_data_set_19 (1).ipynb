```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example DataFrame creation
data = {'Time': [1, 2, 3, 4, 5],
        'Country': ['A', 'B', 'C', 'A', 'B']}
df = pd.DataFrame(data)

# Set up the figure with subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 8))

# Plot histogram for 'Time' column
sns.histplot(df['Time'], ax=axes[0], color='blue')
axes[0].set_title('Distribution of Time')

# Plot histogram for 'Country' column
sns.histplot(df['Country'], ax=axes[1], color='green')
axes[1].set_title('Distribution of Country')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example DataFrame creation
data = {'Time': [1, 2, 3, 4, 5],
        'Country': ['A', 'B', 'C', 'A', 'B']}
df = pd.DataFrame(data)

# Map countries to numerical values
country_mapping = {'A': 0, 'B': 1, 'C': 2}
df['Country_Num'] = df['Country'].map(country_mapping)

# Plot Time vs Country_Num
plt.scatter(df['Time'], df['Country_Num'], color='red')
plt.xlabel('Time')
plt.ylabel('Country (Encoded)')
plt.title('Scatter Plot of Time vs Country (Encoded)')
plt.yticks(ticks=[0, 1, 2], labels=country_mapping.keys())  # Set y-tick labels to original country names
plt.grid(True)
plt.show()
```
